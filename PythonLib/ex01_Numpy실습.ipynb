Numpy
- 파이썬에서 다차원 배열, 행렬 데이터 구조를 제공하는 과학계산에 특화된 라이브러리(모듈)
- 동일한 자료형을 가지는 값들이 배열의 형태로 존재
- 데이터 접근을 최적화하기 위해 인덱스 부여
    ==> 인덱싱, 슬라이싱
- 배열 생성
    1차원 : np.array(단일 리스트)
            np.arange(시작 값, 끝 값, 증감 값)
            --> 1차원으로만 생성
    2차원 : np.array(이중 리스트)
            arange() 함수를 통해 만들어진 1차원 배열을 2차원으로 바꾸고자 할 떄
            --> .reshape(행, 열)
- 배열 속성 확인
    크기 : shape
    차원 : ndim
    요소개수 : size
    타입 : dtype
    
- 데이터 접근
    - 인덱스 기반으로 접근 : 인덱싱, 슬라이싱
    - 특정 기준에 의해서 접근 : 불리언 인덱싱
        --> 논리 연산자를 활용하여 True, False 결과르 출력 후 True에 해당하는 데이터만 접근(False값은 출력하지 않음)
            cf. 두 가지 조건에 대해 비교할때 비트논리 연산자 : &, |
            
- 여러 요소의 데이터에 접근할 때는 [] 한 번더 써줘야 함.
    ex) 1차원 : arr[[0,2]]
        2차원 : arr2[행][열], arr2[행, 열], arr2[시작행:끝행, 시작열:끝열], arr2[행, [열1, 열2]]
        
- 연산 : 배열 요소의 위치적으로 대응된 배열 간 연산 수행

#### 조건에 맞는 데이터 접근하기
- 불리언 인덱싱

```python
# random.choice : 랜덤으로 하나의 요소를 취득
# random.sample : 랜덤으로 여러 요소를 취득, 중복 없음
# random.choices : 랜덤으로 여러 요소를 취득, 중복 있음
```

```python
# 불리언 배열 생성
sample_arr = [True, False]

np.random.seed(1)
bool_arr = np.random.choice(sample_arr, size = 10)
display(bool_arr)

tmp_data = np.arange(10) # 0 ~ 9까지 1씩 증가하는 숫자 생성
display(tmp_data)
```

```python
tmp_data[bool_arr]
```

```python
# 성적데이터 생성 후 조건에 맞는 데이터 접근하기

# 랜덤 값 고정 : 7
np.random.seed(7)
score = np.random.randint(50, 100, size = 20)
score
```

```python
# 논리 연산자를 활용하여 80점 이상인 성적 확인
score[score >= 80]
```

```python
# 70점 미만인 데이터는 몇 개일까?
score[score < 70].size
```

### 영화 평점 데이터 분석- 각 사용자별 평점 평균 구하기

#### 데이터 불러오기
- data 변수에 담기
- np.loadtxt('경로', 구분자, 데이터타입)

```python
data = np.loadtxt('./data/ratings.dat', delimiter='::', dtype = np.int64) # int64 = 2^64
```

```python
print(data)
```

#### 데이터 정보 확인하기

```python
# 차원
print(data.ndim)
# 크기
print(data.shape)
# 요소 개수
print(data.size)
print(data.dtype)
```

#### 전체 평점 평균 구하기

```python
rating_all = data[:, 2].mean()
rating_all
```

```python
# 1. 사용자 아이디만 접근
user_ids = data[:, 0]
user_ids
```

```python
# 2. 사용자 아이드를 중복없이 가져오기
user_id = np.unique(user_ids)
user_id
```

```python
# 3. 사용자 아이디가 1인 평점의 평균 구하기
user_id1 = data[data[:, 0] == 1]
user_id1
```

```python
user_id1[:, 2].mean()
```

```python
avg = []
for i in user_id:
    user_id2 = data[data[:, 0] == i]
    avg.append([i, user_id2[:,2].mean()])
```

```python
avg
```

```python
for i in range(0, len(avg)):
    print(f'ID {i+1}번의 평점 평균 : {avg[i]}')
```

```python
# csv 파일로 저장
np.savetxt('./data/user_rating_mean.csv', avg, delimiter=',', fmt = '%.3f')
```
