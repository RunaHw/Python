### Numpy 배열 다루기

#### 1D array 생성

```python
# 1차원 생성해보기
list1 = [1,2,3,4,5] # 리스트
arr = np.array(list1) # np.array() : 배열을 생성하는 함수
arr # 배열
```

#### 2D array 생성

```python
# 2차원 생성해보기
arr2 = np.array([[1,2,3], [4,5,6]])
arr2
```

#### 배열 속성 확인하기

```python
# 1. 배열의 크기확인 : .shape
print('행 : ', arr2.shape[0], '열 : ' ,arr2.shape[-1])

# 2. 배열의 차원 수 확인 : .ndim
print(arr.ndim)
print(arr2.ndim)

# 3. 배열의 전체 요소 개수 확인 : .size
print(arr.size)
print(arr2.size)

# 4. 배여 데이터 타입 확인 : .dtype
print(arr.dtype)
print(arr2.dtype)
```

#### 다양한 배열 생성 함수들 확인하기
- np.arange() : 사용자가 정의한 숫자만큼의 1차원 배열 생성
- np.arange(시작 값, 끝 값, 증감 값)

```python
# np.arange() 함수를 몰랐을 경우
# 1 ~ 50이 담긴 1차원 배열
list1 = []
for i in range(1, 51) :
    list1.append(i)
np.array(list1)
```

```python
# np.arange() 함수를 사용할 경우
range_arr = np.arange(1,51)
range_arr
```

```python
# 1 ~ 50 담은 5행 10열의 2차원 배열 생성하기
# 1 ~ 50까지 당므 1차원 배열 생성한 후 크기 조정 : reshape
range_arr2 = np.arange(1,51).reshape(5,10)
range_arr2
```

#### 조건에 맞는 데이터 접근하기
- 불리언 인덱싱

```python
# random.choice : 랜덤으로 하나의 요소를 취득
# random.sample : 랜덤으로 여러 요소를 취득, 중복 없음
# random.choices : 랜덤으로 여러 요소를 취득, 중복 있음
```

```python
# 불리언 배열 생성
sample_arr = [True, False]

np.random.seed(1)
bool_arr = np.random.choice(sample_arr, size = 10)
display(bool_arr)

tmp_data = np.arange(10) # 0 ~ 9까지 1씩 증가하는 숫자 생성
display(tmp_data)
```

```python
tmp_data[bool_arr]
```

```python
# 성적데이터 생성 후 조건에 맞는 데이터 접근하기

# 랜덤 값 고정 : 7
np.random.seed(7)
score = np.random.randint(50, 100, size = 20)
score
```

```python
# 논리 연산자를 활용하여 80점 이상인 성적 확인
score[score >= 80]
```

```python
# 70점 미만인 데이터는 몇 개일까?
score[score < 70].size
```

```python
# 성적이 60점 이상인 학생들의 번호 출력하기
score >= 60
num = np.arange(1,21)
num
num[score >= 60]
```
