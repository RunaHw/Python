```python
cnt_mask = int(input('마스크 개수 입력 >> '))



if True :
    print('8개들이 포장지 개수 : ', cnt_mask//8)
    if(cnt_mask % 8) % 5 > 0 :
        print('5개들이 포장지 개수 : ', ((cnt_mask%8) //5)+1)
    else :
        print('5개들이 포장지 개수 : ', (cnt_mask%8) //5)
```

```python
cnt_mask = int(input('마스크 개수 입력 >> '))

# 필요한 변수 선언
box8, box5, temp = 0, 0, 0

# 포장지 개수
box8 = cnt_mask // 8 # 8개 묶음 포장의몫
box5 = cnt_mask %8 //5 # 5개 묶음 포장의 몫
temp = cnt_mask %8 %5 # 5개들이 포장지에 넣고 남은 나머지 

if temp > 0 :
    box5 += 1
    
print('8개들이 포장지 개수 : ', box8)
print('5개들이 포장지 개수 : ', box5)
```

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

if num1 == num2:
    print('두 수가 똑같습니다.')
elif num1 > num2 :
    print('첫 번째 정수가 더 큽니다.')
else :
    print('두 번째 정수가 더 큽니다.')
```

```python
score = int(input('점수 입력 >> '))
if score <= 100 :
    if score >= 90 :
        print(f'{score}점은 A학점 입니다.')
    elif score >= 80 :
        print(f'{score}점은 B학점 입니다.')
    elif score >= 70 :
        print(f'{score}점은 C학점 입니다.')
    elif score >= 60 :
        print(f'{score}점은 D학점 입니다.')
    else :
        print(f'{score}점은 F학점 입니다.')
else : 
    print('점수를 잘 못 입력하였습니다.')
```
