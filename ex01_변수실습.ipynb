# 주석
- 여러줄의 코드를 주석처리 : ctrl+ /
- 한줄 코드를 주석 처리 : shift +  ==> #
- 파란색 : command mode ==> 셀 자체를 설정
- 초록색 : eidt mode ==> 소스 코드 작성

# command mode
- a : 위로 셀 추가
- b : 아래로 셀 추가
- dd : 셀 삭제
- m : 마크다운 전환(메모장처럼 글 작성 가능, 캡쳐한 이미지 넣기도 가능)
- y : 상태를 뒤로 되돌리기
- l : 줄 번호 생성

# 실행 단축키 : 2가지
- ctrl + enter : 해당 셀 하나만 실행
- shift + enter, Alt + enter : 셀 실행하고 아래로 커서 이동(자동으로 셀이 추가)

# edit mode
- esc : command mode로 전환
- ctrl + c : 복사
- ctrl + v : 붙여넣기
- ctrl + x : 잘라내기
- ctrl + z : 앞으로 되돌리기
- ctrl + y : 뒤로 되돌리기
- ctrl + d : 커서행 삭제

- format 함수를 사용한 포맨팅
- f문자열 포매팅

## %기호 포매팅


```python
year = 2023
month = 6
day = 23
s = '오늘은 %d년 %d월 %d일 입니다.'%(year, month, day)
s
```

```python
month = 6
day = 22
s = '오늘은 {}월 {}일 입니다.'.format(month, day)
s
```

```python
lang = "'python'"
alpa = f'Life is too short, You need {lang}'
print(alpa)
```

```python
lang = "'python'"
alpa = 'Life is too short, You need %s'%(lang)
print(alpa)
```
