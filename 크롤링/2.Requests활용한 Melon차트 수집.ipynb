# 코스피 시가총액 TOP50기업 수집

```python
res = req.get("https://finance.naver.com/sise/sise_market_sum.naver")
```

```python
res.text
```

```python
soup = bs(res.text, "lxml")
```

```python
#기업의 이름
title = soup.select(".tltle")
```

```python
# 현재가
price = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr> td:nth-child(3)")
```

```python
title_list = []
price_list = []
rank_list = []
for i in range(len(title)):
    title_list.append(title[i].text)
    price_list.append(price[i].text)
    rank_list.append(i+1)
    
len(title_list)
len(price_list)
```

```python
sise = {"순위" : rank_list,"기업명":title_list, "금액":price_list}
sise_table= pd.DataFrame(sise)
sise_table.set_index("순위", inplace=True)
sise_table
```

```python
sise_table.to_html("코스피TOP50.html")
```
