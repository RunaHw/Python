# Beautifulsoup ÎùºÏù¥Î∏åÎü¨Î¶¨
- reqÎ°ú ÎÑòÍ≤®Î∞õÏùÄ Îç∞Ïù¥ÌÑ∞(ÌÖçÏä§Ìä∏)Î•º Ïª¥Ìì®ÌÑ∞Í∞Ä Ïù¥Ìï¥Ìï† Ïàò ÏûàÍ≤å htmlÌòïÏãùÏúºÎ°ú Î≥ÄÌòï
- beautifulsoup Í∞ùÏ≤¥Ìôî

```python
from bs4 import BeautifulSoup as bs
```

```python
# bs("Î≥ÄÌòïÌï† ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞", "lxml")
# ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Í∞Ä Ïª¥Ìì®ÌÑ∞Í∞Ä Ïù¥Ìï¥Ìï† Ïàò ÏûàÎäî ÌòïÌÉúÎ°ú Î≥ÄÌòï > ''ÏÇ¨ÎùºÏßê
soup = bs(res.text, "lxml")
```

```python
# soup Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÌäπÏ†ï 'ÏöîÏÜå'Î•º ÏàòÏßë
# soup.select("ÏöîÏÜåÏùò Íµ¨Î∂ÑÏûê") > ÌäπÏ†ï ÏöîÏÜåÎßå ÏàòÏßë
# select vs select_one = Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÏßë Í∞úÏàòÏ∞®Ïù¥ Î≥µÏàò, Îã®Ïàò
news_title = soup.select("a.news_tit")
```

```python
soup.select_one("a.news_tit")
```

```python
news_title[0].text
```

```python
# Î™®Îì† Îâ¥Ïä§Ïù¥ ÌÉÄÏù¥ÌãÄÏùÑ Ï∂úÎ†•
for i in news_title:
    print(i.text)
```

```python
text_title = []
for i in news_title:
    text_title.append(i.text)
text_title
```

# ÏÉÅÏö∞Ï¶ùÍ∂å Ï¢ÖÎ™© ÏàòÏßë

```python
res = req.get("https://finance.naver.com/")
```
