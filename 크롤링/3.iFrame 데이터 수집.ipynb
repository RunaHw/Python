```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res = req.get('https://finance.naver.com/marketindex/exchangeList.naver')
res
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
hwapye = soup.select('td.tit > a')
len(hwapye)
hwapye
```

```python
trading = soup.select('td.sale')
len(trading)
```

```python
rank_list = []
hwapye_list = []
trading_list = []
for i in range(len(trading)):
    rank_list.append(i+1)
    hwapye_list.append(hwapye[i].text)
    trading_list.append(trading[i].text)
```
