### 산술 연산자

#### 숫자 연산


```python
num1 = 10
num2 = 7
print('num1/num2 = %0.2f'%(num1/num2))
print("num1%%num2 = %d"%(num1%num2))
print('num1//num2 = %d'%(num1//num2))
```

#### 숫자, 문자열을 더했을 때

```python
num1 = 10  # 정수 자료형
str1 = '7' # 문자열 자료형

# print(num1 + str2)
# 형변환 : (int)x ==> int(대상)
print(num1 + int(str1))
print(str(num1) + str1)
```

```python
num1 = int(input("정수를 입력하세요 >> "))
num2 = int(input("정수를 입력하세요 >> "))

print("더하기 결과 : %d"%(num1+num2))
print("빼기 결과 : %d"%(num1-num2))
print("곱하기 결과 : %d"%(num1*num2))
print("나누기 결과 : %f"%(num1/num2))
```

### 비교 연산자
- True / False로 결과가 출력되는 연산자

  : >, >=, <. <=, ==, !=

```python
a = 3
b = 7
print(a>b)
print(a==b)
print(a!=b)
print(a<b)
```

### 논리 연산자
- not : 논리를 뒤집는 연산자 True -> False, False -> True
- and, or

```python
# not
a, b = 5, 10
not a < b # a < b : True --> not ==> False
```

### 멤버 연산자

```python
# not in 
'우리반' not in str1
```

### 삼항 연산자
- True 실행문장 if 조건식 else False 실행문장

```python
# 60점 이상이면 합격, 미만이면 불합격 출력하기
score = 50
'합격'if score >= 60 else '불합격'
```

```python
a = int(input('정수입력 >> '))
b = int(input('정수입력 >> '))

print(f"a : {a}" if a > b else f"b : {b}")
```

```python
num = int(input('정수 입력 >> '))
f"{num}는(은)짝수입니다."if (num%2)==0 else f"{num}는(은)홀수입니다."
```
